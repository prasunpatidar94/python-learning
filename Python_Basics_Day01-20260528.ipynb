#### Python DataTypes:
- Premative : int, double, complex, str, float, boolean
- Container Type: List, Set, Tuple, Dictionary
- User Defined: class  type




```
# Int (premetive)
intA= 10
print(intA)
print (type(intA))
```

```
# float (premative)
Afloat=3.47
print(Afloat)
print(type(Afloat))
```

```
# boolean (premative)
Abool = True
print(Abool)
print(type(Abool))
```

```
# str (premative)
Astr = "Prasun"
Bstr = 'Prasun'
Cstr = """Prasun"""
Dstr = '''Prasun'''
print(Astr)
print(Bstr)
print(Cstr)
print(Dstr)
print(type(Astr))
```

```
# Complex (premative)
AComplex = 3 + 7j
print(AComplex)
print(type(AComplex))
```

```
# List (Container Type)
AList = ["I","Am","Prasun","Patidar"]
print(AList)
print(type(AList))
```

```
# Tuple (Container Type)
ATupe = ("I","Am","Prasun","Patidar")
print(ATupe)
print(type(ATupe))
```

```
# Set (container Type)
ASet = {1,3,4,5,6,7,8,10}
print(ASet)
print(type(ASet))
```

```
# Dic (Container Type)
ADic = {"name":"Prasun","age":32,"work":"Software"}
print(ADic)
print(type(ADic))
```

```
# range (Premative)
ARange = range(10);
print(ARange)
print(type(ARange))
```
