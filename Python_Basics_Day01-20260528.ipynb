#### Python DataTypes:
- Premative : int, double, complex, str, float, boolean
- Container Type: List, Set, Tuple, Dictionary
- User Defined: class  type




```
# Int (premetive)
intA= 10
print(intA)
print (type(intA))
```

```
# float (premative)
Afloat=3.47
print(Afloat)
print(type(Afloat))
```

```
# boolean (premative)
Abool = True
print(Abool)
print(type(Abool))
```

```
# str (premative)
Astr = "Prasun"
Bstr = 'Prasun'
Cstr = """Prasun"""
Dstr = '''Prasun'''
print(Astr)
print(Bstr)
print(Cstr)
print(Dstr)
print(type(Astr))
```
